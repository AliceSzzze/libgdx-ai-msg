```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
def stats(num):
    df = pd.read_csv('cmake-build-debug/exp'+str(num)+'.csv', skiprows=6)
    print(df.describe())
    # Plot histogram using pandas
    df[df['latency'] > 1]['latency'].plot(kind='hist', bins=5, edgecolor='black')

    # Add labels and title
    plt.xlabel('latency (ms)')
    plt.ylabel('Frequency')
    plt.title('distribution of latency > 1 ms')

    # Show the plot
    plt.show()
    
    print(str(len(df[df['latency'] > 1])/len(df)*100) +"% of sends have > 1ms latency")
    print(str(len(df[df['latency'] > 16])/len(df)*100) +"% of sends have > 16ms latency")
```
